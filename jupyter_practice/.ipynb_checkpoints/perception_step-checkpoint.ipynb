```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Read in sample image
image_name = '/home/dell/udacity_projects/jupyter_practice/img2017-03-26-20-25-10-378.jpg'
image = mpimg.imread(image_name)
```

```python
# Define a function to perform a color threshold
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    color_select = np.zeros_like(img[:, :, 0])
    for h, y in enumerate(img):
        for w, x in enumerate(y):   
            if x[0] > rgb_thresh[0] and x[1] > rgb_thresh[1] and x[2] > rgb_thresh[2]:
                color_select[h, w] = 1
        else:
            color_select[h, w] = 0
            
    return color_select
```

```python
# Define color selection criteria
red_threshold = 160
green_threshold = 160
blue_threshold = 160

rgb_threshold = (red_threshold, green_threshold, blue_threshold)

# pixels below the thresholds
colorsel = color_thresh(image, rgb_thresh=rgb_threshold)

# Display the original image and the binary
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(21,7), sharey=True)
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=40)

ax2.imshow(colorsel, cmap='gray')
ax2.set_title('Your Result', fontsize=40)

plt.subplots_adjust(left=0.0, right=1, top=0.9, bottom=0.0)
plt.show()
```
